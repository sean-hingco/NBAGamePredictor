```python
import pandas as pd
import numpy as np
```

```python
data = '../data/merged_stats_22024.csv'
df = pd.read_csv(data)
df.head()
```

```python
# normalization
from sklearn.preprocessing import StandardScaler, MinMaxScaler

numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
exclude_cols = ['GAME_ID', 'PLAYER_ID', 'TEAM_ID', 'jerseyNum']
numeric_cols = [col for col in numeric_cols if col not in exclude_cols]

scaler = StandardScaler()

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
df[numeric_cols].describe()
```

```python
# encode position
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['position_encoded'] = label_encoder.fit_transform(df['position'])
df['position_encoded']
```

```python
# convert minutes and date to numeric values
def time_to_minutes(time_str):
    if pd.isna(time_str):
        return np.nan
    minutes, seconds = time_str.split(':')
    return int(minutes) + int(seconds) / 60

df['minutes'] = df['minutes'].apply(time_to_minutes)

df['GAME_DATE'] = pd.to_datetime(df['GAME_DATE'], errors='coerce')
df['days_since'] = (df['GAME_DATE'] - df['GAME_DATE'].min()).dt.days

df[['minutes', 'days_since']]
```

```python
# change WL to numeric
df['WL'] = df['WL'].map({'W': 1, 'L': 0})
df['WL'].head()
```

```python
# encode opponent
df['opposing_team'] = df['MATCHUP'].str.extract(r'(\w+)\s*(vs\.|@)\s*(\w+)')[2]

team_mapping = {team: idx for idx, team in enumerate(df['opposing_team'].unique())}
df['opposing_team_numeric'] = df['opposing_team'].map(team_mapping)

df.drop(['opposing_team'], axis=1, inplace=True)

df[['MATCHUP', 'opposing_team_numeric']].head()
```

```python
# drop unnecessary columns
non_numeric_columns = df.select_dtypes(exclude=['number']).columns.tolist()
df = df.drop(columns=non_numeric_columns)

df.columns
```

```python
# check for nans or infs
print("NaN values in dataset:\n", np.isnan(df[numeric_cols]).sum())  # Count NaNs
print("Inf values in dataset:\n", np.isinf(df[numeric_cols]).sum())  # Count Inf values
```

```python
# save normalized data
df.to_csv('preprocessed_stats_22024.csv', index=False)
```
